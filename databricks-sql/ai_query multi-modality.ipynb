```sql
%python

import os
from PIL import Image
from IPython.display import display
from pathlib import Path

# Path to the volume directory
image_dir = "/Volumes/main/kent_dais_demo/km"

# Supported image extensions
image_extensions = [".png"]

# List image files in the directory
image_files = [f for f in os.listdir(image_dir) if Path(f).suffix.lower() in image_extensions]

# Display each image
for img_file in image_files:
    img_path = os.path.join(image_dir, img_file)
    try:
        img = Image.open(img_path)
        print(f" Showing: {img_file}")
        display(img)
    except Exception as e:
        print(f" Failed to display {img_file}: {e}")
```

```sql
%sql

SELECT path, 
  AI_QUERY("databricks-llama-4-maverick", "What is interesting or engaging about this image?", 
    responseFormat => "
      STRUCT<
        output:STRUCT<
          summary: STRING ,
          interesting_fact: STRING ,
          possible_context:STRING ,
          num_people: INTEGER,
          num_animals: INTEGER
        > 
      >
    ",
    files => content
  ) as output FROM READ_FILES("/Volumes/main/kent_dais_demo/km/*.png")
```

### Write a caption for a social media post.

```sql

%sql
SELECT path, 
  AI_QUERY("databricks-llama-4-maverick", "If this image were used in a presentation or social media post, what would a provocative caption be?", 
    responseFormat => "
      STRUCT<
        output:STRUCT<
          caption: STRING COMMENT 'Short, catchy, and relevant text for the image.',
          tone: STRING COMMENT 'e.g. professional, playful, inspiring',
          hashtags: ARRAY<STRING>   
        > 
      >
    ",
    files => content
  ) as output FROM READ_FILES("/Volumes/main/kent_dais_demo/km/*.png")

```

### What inferences can you make about the situation, intent, or relationships between the subjects?

```sql
%sql

SELECT path, 
  AI_QUERY("databricks-claude-3-7-sonnet", "Based on the image, what inferences can you make about the situation, intent, or relationships between the subjects?", 

    files => content,
    modelParameters =>
      named_struct(
        "max_tokens", 2000, "thinking", named_struct("type", "enabled", "budget_tokens", 1024)
      )
  ) as output FROM READ_FILES("/Volumes/main/kent_dais_demo/km/*.png")
```
